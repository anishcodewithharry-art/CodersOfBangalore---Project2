```python
with open("initialdata.txt", encoding='utf-8') as file:
    data = file.read()
```

```python
data
```

```python
chunks = data.split("\n\n")
chunks = [c for c in chunks if len(c)>3]
```

```python
chunks[5]
```

```python
# Function to parse the chunks one-by-one
def parse_chunk(chunk):
    try:
        chunk = chunk.strip()
        splitted_chunk= chunk.split("\n")
    
        username = splitted_chunk[0]
        no_of_posts = int(splitted_chunk[1].split(" post")[0].replace(",", ""))
        no_of_followers = float(splitted_chunk[2].split(" followers")[0].replace(",", "").replace("K", "").replace("M", ""))

        if "K" in splitted_chunk[2].split(" followers")[0]:
            no_of_followers = int(no_of_followers * 1000)
        elif "M" in splitted_chunk[2].split(" followers")[0]:
            no_of_followers = int(no_of_followers * 1000000)
        else:
            no_of_followers = int(no_of_followers)
            
        # no_of_following = int(splitted_chunk[3].split(" following")[0].replace(",", ""))

        no_of_following = float(splitted_chunk[3].split(" following")[0].replace(",", "").replace("K", "").replace("M", ""))

        if "K" in splitted_chunk[3].split(" following")[0]:
            no_of_following = int(no_of_following * 1000)
        elif "M" in splitted_chunk[3].split(" following")[0]:
            no_of_following = int(no_of_following * 1000000)
        else:
            no_of_following = int(no_of_following)
        
        name = splitted_chunk[4]

        if (len(splitted_chunk)>5):
            type_of_page = splitted_chunk[5]
            bio = splitted_chunk[6:]   
        else:
            type_of_page = "Unknown"
            bio = "Unknown"
    except Exception as e:
        print(e)
        print(chunk)
    # print(username, no_of_posts, no_of_followers, no_of_following, name)

    # print(splitted_chunk[2].split(" followers")[0])
    # print(no_of_followers)

    return {"username": username, 
             "no_of_posts": no_of_posts,
             "no_of_followers": no_of_followers,
             "no_of_following": no_of_following,
             "name": name,
             "type_of_page": type_of_page,
             "bio": bio
           }
```

```python
all_chunks = []
for chunk in chunks:
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)

# print(all_chunks)

# parse_chunk(chunk)
```

```python
import json

with open("initialdata.json", "w") as file:
    json.dump(all_chunks, file, indent=4)
```

```python
with open("initialdata.json", "r") as file:
    data = json.load(file)

data
```

```python
# find person with maximum Posts

max_post = 0
for chunk in all_chunks:
    if (max_post < chunk["no_of_posts"]):
        max_post = chunk["no_of_posts"]
        chunk_with_max_posts = chunk

print(chunk_with_max_posts)
```
